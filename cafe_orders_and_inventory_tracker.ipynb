```python
menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25
}

orders = [
    {"tea": 2, "cookie": 1},
    {"coffee": 1, "sandwich": 2},
    {"tea": 3, "cake_slice": 1},
    {"coffee": 2, "cookie": 2, "sandwich": 1},
    {"tea": 1},
    {"coffee": 1, "cake_slice": 2}
]
```

```python
processed_orders = []

for order in orders:
    clean_order = {}
    for item, qty in order.items():
      if item not in menu:
            continue

      available = inventory[item]
      if qty > available:
            qty = available

      if qty > 0:
            clean_order[item] = qty

    processed_orders.append(clean_order)
```

```python
order_totals = []

for order in processed_orders:
    total = 0
    for item, qty in order.items():
        total += qty * menu[item]
    order_totals.append(total)
```

```python
items_sold = {}

for order in processed_orders:
    for item, qty in order.items():
        inventory[item] -= qty
        items_sold[item] = items_sold.get(item, 0) + qty
```

```python
total_revenue = sum(order_totals)
top_selling_item = max(items_sold, key=items_sold.get)
```

```python
low_stock = [item for item, qty in inventory.items() if qty <= 5]
```

```python
order_totals_with_discount = []
for total in order_totals:
    if total > 1000:
        total = total * 0.90   # 10% discount
    order_totals_with_discount.append(total)

discounted_revenue = sum(order_totals_with_discount)
```

```python
order_points = []
for total in order_totals_with_discount:
    points = total // 100
    order_points.append(int(points))

total_points_given = sum(order_points)
```

```python
print("===== PROCESSED ORDERS =====")
print(processed_orders)

print("\n===== ORDER TOTALS =====")
print(order_totals)

print("\n===== INVENTORY AFTER SALES =====")
print(inventory)

print("\n===== DAY SUMMARY =====")
print("Total Revenue (without discount):", total_revenue)
print("Items Sold:", items_sold)
print("Top Selling Item:", top_selling_item)

print("\n===== LOW STOCK ITEMS (<=5 LEFT) =====")
print(low_stock)

print("\n===== ORDER TOTALS WITH DISCOUNT =====")
print(order_totals_with_discount)
print("Total Revenue (with discount):", discounted_revenue)

print("\n===== LOYALTY POINTS =====")
print("Points per order:", order_points)
print("Total Points Given Today:", total_points_given)
```
